```
#Hadamard gate
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister,Aer,execute
from qiskit.visualization import plot_bloch_multivector,plot_histogram,iplot_bloch_multivector

qr = QuantumRegister(2)
cr = ClassicalRegister(2)
circuit = QuantumCircuit(qr,cr)
circuit.h(1) #harmard gate
circuit.cx(0,1) #control not gates
circuit.draw(output='mpl')
```

```
result = execute(circuit,Aer.get_backend('statevector_simulator'),shots = 1024).result()
statevec = result.get_statevector()
plot_bloch_multivector(statevec)
```
