```
from qiskit import QuantumCircuit,Aer,execute
from qiskit.visualization import plot_histogram
from matplotlib import pyplot

circuit = QuantumCircuit(3,3)
circuit.x(0)
circuit.barrier()
circuit.h(1)

circuit.cx(1,2)
circuit.draw(output='mpl')
```

```
circuit.barrier()
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```
circuit.barrier()
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

```
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```
#let verify
circuit.measure(2,2)
result = execute(circuit,backend =Aer.get_backend('qasm_simulator'),shots=1024).result()
plot_histogram(result.get_counts())
```

```
circuit.measure(2,2)
result = execute(circuit,backend =Aer.get_backend('statevector_simulator'),shots=1024).result()
statenv = result.get_statevector()
from qiskit.visualization import plot_bloch_multivector
plot_bloch_multivector(statenv)
```
